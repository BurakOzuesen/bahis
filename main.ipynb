```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/2018-2019.csv")
df.head()
```

```python
df.columns
```

```python
df_redined = df.iloc[:, 2:10]
df_redined.columns
```

```python
df_redined.head()
```

```python
# df_redined["Home_Starting_Elo"] = ""
# df_redined["Home_Ending_Elo"] = ""
# df_redined["Away_Starting_Elo"] = ""
# df_redined["Away_Ending_Elo"] = ""
```

```python
df_encoded = pd.get_dummies(df_redined)
```

```python
df_encoded
```

```python
X = df_encoded.iloc[:, 4:44]
y = df_encoded.iloc[:, 44:47]
y
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2)
```

```python
test_y
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
rfc.fit(train_X, train_y)
```

```python
preds = rfc.predict(test_X)
```

```python
preds
```

```python
test_y
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_y, preds)
```
